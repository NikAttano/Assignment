## Here I am using wratio described here:
https://towardsdatascience.com/fuzzywuzzy-fuzzy-string-matching-in-python-beginners-guide-9adc0edf4b35

For performing string matching between input and the 2 datasets

```python
x = input('Enter the product for example try "WALLDESIGN SMALL" : ')
amazon_score = 0
flipkart_score = 0
max_amazon_score = 0
max_flipkart_score = 0
amazon_product = ''
flipkart_product = ''
amz_index = 0
fk_index = 0
amz_price = 0
fk_price = 0
amz_dis_price = 0
fk_dis_price = 0

for i in range(df.shape[0]):
  amazon_score = fuzz.WRatio(df['product_name_amazon'][i], x)
  if amazon_score > max_amazon_score:
    max_amazon_score = amazon_score
    amazon_product = df['product_name_amazon'][i]
    amz_index = i
for i in range(df.shape[0]):
  flipkart_score = fuzz.WRatio(df['product_name_flipkart'][i], x)
  if flipkart_score > max_flipkart_score:
    max_flipkart_score = flipkart_score
    flipkart_product = df['product_name_flipkart'][i]
    fk_index = i


  
print('Amazon product found as :', amazon_product)
print('Flipkart product found as :', flipkart_product)
fk_price = df1['retail_price'][fk_index]
fk_dis_price = df1['discounted_price'][fk_index]
amz_price = df2['retail_price'][amz_index]
amz_dis_price = df2['discounted_price'][amz_index]
df_display = pd.DataFrame()
df_display.insert(0, 'Product name in Flipkart', flipkart_product)
df_display.insert(1, 'Retail Price in Flipkart', fk_price)
df_display.insert(2, 'Discounted Price in Flipkart', fk_dis_price)
df_display.insert(3, 'Product name in Amazon', amazon_product)
df_display.insert(4, 'Retail Price in Amazon', amz_price)
df_display.insert(5, 'Discounted Price in Amazon', amz_dis_price)
# df_display.loc[len(df.index)] = ['Product name in Flipkart', 'Retail Price in Flipkart', 'Discounted Price in Flipkart', 'Product name in Amazon', 'Retail Price in Amazon', 'Discounted Price in Amazon']
df_display.loc[len(df_display.index)] = [flipkart_product, fk_price, fk_dis_price, amazon_product, amz_price, amz_dis_price]
df_display
```

## Here I am passing all 20,000 sentences through the model to generate embeddings, that is the sentence gets converted to a tensor of embedding values.
It gets finally stored in a dataframe df_emb

```python
# This cell takes 15 minutes to run and generates a csv with the word embeddings for all 20,000 sentences
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer(model_name)

df_emb = pd.DataFrame(columns = ['amazon_embeddings', 'flipkart_embeddings'])
df_emb['amazon_embeddings'] = df_emb['amazon_embeddings'].astype(object)
df_emb['flipkart_embeddings'] = df_emb['flipkart_embeddings'].astype(object)
fk_price = df1['retail_price']
fk_dis_price = df1['discounted_price']
amz_price = df2['retail_price']
amz_dis_price = df2['discounted_price']
iter_list = range(0, df.shape[0])
for i in iter_list:
  amz_emb = model.encode(df['product_name_amazon'][i], convert_to_tensor=True)
  fk_emb = model.encode(df['product_name_flipkart'][i], convert_to_tensor=True)
  df_emb.loc[i, 'amazon_embeddings'] = amz_emb.numpy()
  df_emb.loc[i, 'flipkart_embeddings'] = fk_emb.numpy()
df_emb.insert(df_emb.shape[1], 'FK_prices', fk_price)
df_emb.insert(df_emb.shape[1], 'FK_dis_prices', fk_dis_price)
df_emb.insert(df_emb.shape[1], 'amz_prices', amz_price)
df_emb.insert(df_emb.shape[1], 'amz_dis_prices', amz_dis_price)
df_emb.insert(df_emb.shape[1], 'amz_name', df['product_name_amazon'])
df_emb.insert(df_emb.shape[1], 'fk_name', df['product_name_flipkart'])
df_emb.head()
df_emb.to_csv('/content/sample_data/embeddings_df.csv', index = True)
```

## Here I am passing the input to the model to generate embeddings.
Finally I am taking the input embedding and comparing with embeddings previously generated.
Also I am using cosine similarity to compare the two embeddings to find the maximum matched sentences.

```python
from re import I
import torch
import numpy as np
amz_score = 0
amz_ind = 0
x = input('Enter the product for example try "Model" : ')
inp_emb = model.encode(x, convert_to_tensor=True)
count = 0
for i in range(df_emb['amazon_embeddings'].shape[0]):
  e = df_emb['amazon_embeddings'][i]  
  sim_score = util.pytorch_cos_sim(inp_emb, torch.from_numpy(e))
  sim_score = sim_score.numpy()[0]
  if sim_score > amz_score:
    amz_score = sim_score
    amz_ind = i

amazon_product = df_emb['amz_name'][amz_ind]
flipkart_product = df_emb['fk_name'][amz_ind]


print('Amazon product found as :', amazon_product)
print('Flipkart product found as :', flipkart_product)
fk_price = df_emb['FK_prices'][amz_ind]
fk_dis_price = df_emb['FK_dis_prices'][amz_ind]
amz_price = df_emb['amz_prices'][amz_ind]
amz_dis_price = df_emb['amz_dis_prices'][amz_ind]
df_display_1 = pd.DataFrame()
df_display_1.insert(0, 'Product name in Flipkart', flipkart_product)
df_display_1.insert(1, 'Retail Price in Flipkart', fk_price)
df_display_1.insert(2, 'Discounted Price in Flipkart', fk_dis_price)
df_display_1.insert(3, 'Product name in Amazon', amazon_product)
df_display_1.insert(4, 'Retail Price in Amazon', amz_price)
df_display_1.insert(5, 'Discounted Price in Amazon', amz_dis_price)
# df_display.loc[len(df.index)] = ['Product name in Flipkart', 'Retail Price in Flipkart', 'Discounted Price in Flipkart', 'Product name in Amazon', 'Retail Price in Amazon', 'Discounted Price in Amazon']
df_display_1.loc[len(df_display_1.index)] = [flipkart_product, fk_price, fk_dis_price, amazon_product, amz_price, amz_dis_price]
df_display_1
```
